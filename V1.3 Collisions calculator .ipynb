```python

def anAlgebra(num1,num2):
    if type(num1)==float:
        result=str(num1)+str(num2)
    elif type(num2)==float:
        result=str(num2)+str(num1)
    return result

def floating(userInput):
    try:
        userInput=float(userInput)
        userInput/2
    except:
        userInput=str(userInput)
    else:
        userInput=float(userInput)
    return userInput

def algebraMultiply(num1,num2):
    if type(num1)==float:
        number=num1
    elif type(num2)==float:
        number=num2
    try:
        test=num1*num2
    except TypeError:
        result=anAlgebra(num1,num2)
    else:
        result=num1*num2
    return result 

def algebraDivide(num1,num2):
    try:
        result=num1/num2
    except TypeError:
        if type(num1)==float:
            result=str(num2)+"/"+str(num1)
        elif type(num2)==float:
            result=str(num1)+"/"+str(num2)
    else:
        result=num1/num2
    return result

def algebraAdd(num1,num2):
    try:
        result=num1+num2
    except TypeError:
        if type(num1)==float:
            result=str(num2)+"+"+str(num1)
        elif type(num2)==float:
            result=str(num1)+"+"+str(num2)
    else:
        result=num1+num2
    return result

def algebraSubtract(num1,num2):
    try:
        result=num1-num2
    except TypeError:
        if type(num1)==float:
            result=str(num2)+"-"+str(num1)
        elif type(num2)==float:
            result=str(num1)+"-"+str(num2)
    else:
        result=num1-num2
    return result
```

```python
initialVelocity1=floating(input("Enter the initial velocity in m/s of particle 1:\n"))

initialVelocity2=floating(input("Enter the initial velocity in m/s of particle 2:\n"))

finalVelocity1=floating(input("Enter the final velocity in m/s of particle 1:\n"))

finalVelocity2=floating(input("Enter the final velocity in m/s of particle 2:\n"))

mass1=floating(input("Enter the mass in kg of particle 1:\n"))

mass2=floating(input("Enter the mass in kg of particle 2:\n"))

coefficientOfRestitution=floating(input("Enter the coefficient of restitution between the particles:\n"))
```

```python
def lawOfRestitutionFormula(initialVelocity1,initialVelocity2,finalVelocity1,finalVelocity2):   #particle 1 and particle 2
    coefficientOfRestitution=algebraDivide((algebraSubtract(finalVelocity2,finalVelocity1)),
                                           algebraSubtract((initialVelocity1,initialVelocity2)))
    #coefficientOfRestitution (written as: e) must be 0<=e<=1
    return coefficientOfRestitution

def momentumFormula(mass,velocity):
    momentum=algebraMultiply(mass,velocity)
    return momentum

def impulseFormula(finalMomentum,initialMomentum):
    impulse=algebraSubtract(finalMomentum,initialMomentum)
    return impulse

def kineticEnergyFormula(mass,velocity):
    kineticEnergy=algebraDivide((algebraMultiply(mass,(algebraMultiply(velocity,velocity)),2)))
    return kineticEnergy

def conservationOfLinearMomentumFormula(initialMomentum1,initialMomentum2,finalMomentum1or2):
    finalMomentumToFind=algebraSubtract(algebraAdd(initialMomentum1,initialMomentum2),finalMomentum1or2) 
    #could be finding momentum of particle 1 or particle 2
    return finalMomentumToFind

def lossOfKineticEnergyFromImpactFormula(initialKineticEnergy1,initialKineticEnergy2,finalKineticEnergy1,finalKineticEnergy2):
    lossOfKineticEnergy=algebraSubtract((algebraAdd(initialKineticEnergy1,initialKineticEnergy2)),
                                        (algebraSubtract(finalKineticEnergy1+finalKineticEnergy2)))
    return lossOfKineticEnergy

def workDoneFormula(componentOfForce,distanceMovedInForceDirection):
    workDone=algebraMultiply(componentOfForce,distanceMovedInForceDirection)
    return workDone

def workDoneAgainstGravityFormula(mass,accelerationDueToGravity,heightRaised):
    workDoneAgainstGravity=algebraMultiply(algebraMultiply(mass,accelerationDueToGravity),heightRaised)
    return workDoneAgainstGravity

def findQmark(variable):
    for i in range (len(variable)):
        containsQmark=False
        if variable[i]=="?":
            containsQmark=True
    return containsQmark
```
